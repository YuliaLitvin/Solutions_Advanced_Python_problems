## <center>–ë—ñ–ª—å—à–µ –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ–≥–æ</center>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è —Ä—è–¥–æ–∫ —Ç–µ–∫—Å—Ç—É –∑ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–∏—Ö —á–∏—Å–µ–ª. –ó –Ω–µ—ó —Ñ–æ—Ä–º—É—î—Ç—å—Å—è –ø–µ—Ä–µ–ª—ñ–∫ —á–∏—Å–µ–ª. –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É –ø—ñ–¥—Ä–∞—Ö—É–Ω–∫—É –∫—ñ–ª—å–∫–æ—Å—Ç—ñ —á–∏—Å–µ–ª, —è–∫—ñ –±—ñ–ª—å—à—ñ –∑–∞ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ–π —É —Ü—å–æ–º—É —Å–ø–∏—Å–∫—É —á–∏—Å–ª–∞.

**–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö**

–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è —Ä—è–¥–æ–∫ —Ç–µ–∫—Å—Ç—É –∑ —Ä–æ–∑–¥—ñ–ª–µ–Ω–∏—Ö –ø—Ä–æ–±—ñ–ª–∞–º–∏ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–∏—Ö —á–∏—Å–µ–ª.

**–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö**

–ü—Ä–æ–≥—Ä–∞–º–∞ –º–∞—î –≤–∏–≤–µ—Å—Ç–∏ –æ–¥–Ω–µ —á–∏—Å–ª–æ ‚Äì –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —Å–ø–∏—Å–∫—É, –±—ñ–ª—å—à–∏—Ö –∑–∞ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ–π.

```python
a = input().split()
count = 0
for i in range(1, len(a)):
    if int(a[i]) > int(a[i - 1]):
        count += 1
print(count)
```

## <center>–î–æ–±—É—Ç–æ–∫ —á–∏—Å–µ–ª</center>
–ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É –¥–ª—è –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è, —á–∏ —î —á–∏—Å–ª–æ–º —Ç–≤–æ—Ä–æ–º –¥–≤–æ—Ö —á–∏—Å–µ–ª –∑ —Ü—å–æ–≥–æ –Ω–∞–±–æ—Ä—É, —â–æ –≤–∏–≤–æ–¥–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç —É –≤–∏–≥–ª—è–¥—ñ –≤—ñ–¥–ø–æ–≤—ñ–¥—ñ ¬´–¢–ê–ö¬ª –∞–±–æ ¬´–ù–Ü¬ª.

**–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö**

–£ –ø–µ—Ä—à–æ–º—É —Ä—è–¥–∫—É –ø–æ–¥–∞—î—Ç—å—Å—è —á–∏—Å–ª–æ n(0<n<1000) ‚Äì –∫—ñ–ª—å–∫—ñ—Å—Ç—å —á–∏—Å–µ–ª —É –Ω–∞–±–æ—Ä—ñ. –£ –Ω–∞—Å—Ç—É–ø–Ω–∏—Ö n —Ä—è–¥–∫–∞—Ö –≤–≤–æ–¥—è—Ç—å—Å—è —Ü—ñ–ª—ñ —á–∏—Å–ª–∞, —â–æ —Å—Ç–∞–Ω–æ–≤–ª—è—Ç—å –Ω–∞–±—ñ—Ä (–º–æ–∂—É—Ç—å –ø–æ–≤—Ç–æ—Ä—é–≤–∞—Ç–∏—Å—è). –ü–æ—Ç—ñ–º —Å–ª—ñ–¥—É—î —Ü—ñ–ª–µ —á–∏—Å–ª–æ, —è–∫–µ —î —á–∏ –Ω–µ —î –¥–æ–±—É—Ç–∫–æ–º –¥–≤–æ—Ö —è–∫–∏—Ö–æ—Å—å —á–∏—Å–µ–ª —ñ–∑ –Ω–∞–±–æ—Ä—É.

**–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö**

–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ ¬´–¢–ê–ö¬ª –∞–±–æ ¬´–ù–Ü¬ª –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ —É–º–æ–≤–∏ –∑–∞–≤–¥–∞–Ω–Ω—è.

```python
nums = []
n = int(input())
res = '–ù–ï–¢'
for i in range(n):
    nums.append(int(input()))
big = int(input())
for i in range(n):
    for j in range(n):
        if i != j:
            if nums[i] * nums[j] == big:
                res = '–î–ê'
print(res)
```

## <center>–ö—Ä–µ–º–Ω—ñ—î–≤–∞ –¥–æ–ª–∏–Ω–∞ üå∂Ô∏èüå∂Ô∏è</center>
–®—Ç—É—á–Ω–∏–π —ñ–Ω—Ç–µ–ª–µ–∫—Ç –ê–Ω—Ç–æ–Ω, —Å—Ç–≤–æ—Ä–µ–Ω–∏–π –ì—ñ–ª—Ñ–æ–π–ª–æ–º, –∑–ª–∞–º–∞–≤ –º–µ—Ä–µ–∂—É —Ä–æ–∑—É–º–Ω–∏—Ö —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫—ñ–≤. –¢–µ–ø–µ—Ä –≤—ñ–Ω –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î —ó—Ö —è–∫ —Å–µ—Ä–≤–µ—Ä–∏ "–ü–µ–≥–æ–≥–æ –¥—É–¥–æ—á–Ω–∏–∫–∞". –î–æ–ø–æ–º–æ–∂—ñ—Ç—å –≤–ª–∞—Å–Ω–∏–∫—É —Ñ—ñ—Ä–º–∏ –∑–Ω–∞–π—Ç–∏ –≤—Å—ñ –∑–∞—Ä–∞–∂–µ–Ω—ñ —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫–∏.

–î–ª—è –∫–æ–∂–Ω–æ–≥–æ —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫–∞ —ñ—Å–Ω—É—î —Ä—è–¥–æ–∫ –∑ –¥–∞–Ω–∏–º–∏, —â–æ —Å–∫–ª–∞–¥–∞—î—Ç—å—Å—è –∑ –º–∞–ª–∏—Ö –ª—ñ—Ç–µ—Ä —ñ —Ü–∏—Ñ—Ä, —ñ —è–∫—â–æ –≤ –Ω—ñ–π –ø—Ä–∏—Å—É—Ç–Ω—î —Å–ª–æ–≤–æ "anton" (–Ω–µ–æ–±–æ–≤'—è–∑–∫–æ–≤–æ –ª—ñ—Ç–µ—Ä–∏, —â–æ –ø–æ—Ä—É—á –ø–æ—Ä—É—á, –≥–æ–ª–æ–≤–Ω–∞ –Ω–∞—è–≤–Ω—ñ—Å—Ç—å –ø–æ—Å–ª—ñ–¥–æ–≤–Ω–æ—Å—Ç—ñ –ª—ñ—Ç–µ—Ä), —Ç–æ —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫ –∑–∞—Ä–∞–∂–µ–Ω–∏–π —ñ –ø–æ—Ç—Ä—ñ–±–Ω–æ –≤–∏–≤–µ—Å—Ç–∏ –Ω–æ–º–µ—Ä —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫–∞, –Ω—É–º–µ—Ä–∞—Ü—ñ—è –ø–æ—á–∏–Ω–∞—î—Ç—å—Å—è –∑ –æ–¥–∏–Ω–∏—Ü—ñ.

**–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö**

–£ –ø–µ—Ä—à–æ–º—É —Ä—è–¥–∫—É –ø–æ–¥–∞—î—Ç—å—Å—è —á–∏—Å–ª–æ n ‚Äì –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫—ñ–≤. –£ –Ω–∞—Å—Ç—É–ø–Ω–∏—Ö n —Ä—è–¥–∫–∞—Ö –≤–≤–æ–¥—è—Ç—å—Å—è —Ä—è–¥–∫–∏, —â–æ –º—ñ—Å—Ç—è—Ç—å –ª–∞—Ç–∏–Ω—Å—å–∫—ñ –º–∞–ª—ñ –ª—ñ—Ç–µ—Ä–∏ —Ç–∞ —Ü–∏—Ñ—Ä–∏, —É –∫–æ–∂–Ω–æ–º—É —Ä—è–¥–∫—É –≤—ñ–¥ 5 –¥–æ 100 —Å–∏–º–≤–æ–ª—ñ–≤.

**–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö**

–ü—Ä–æ–≥—Ä–∞–º–∞ –º–∞—î –≤–∏–≤–µ—Å—Ç–∏ –Ω–æ–º–µ—Ä–∏ –∑–∞—Ä–∞–∂–µ–Ω–∏—Ö —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫—ñ–≤ —á–µ—Ä–µ–∑ –ø—Ä–æ–ø—É—Å–∫. –Ø–∫—â–æ —Ç–∞–∫–∏—Ö —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫—ñ–≤ –Ω–µ–º–∞—î, –Ω—ñ—á–æ–≥–æ –Ω–µ –ø–æ—Ç—Ä—ñ–±–Ω–æ –≤–∏–≤–æ–¥–∏—Ç–∏.

```python
s = int(input())
count = 1
result = []
for i in range(s):
    a = input()
    list = []
    for j in a:
        if j == 'a' and list == []:
            list.append(j)
        elif j == 'n' and list == ['a']:
              list.append(j)
        elif j == 't' and list == ['a', 'n']:
              list.append(j)
        elif j == 'o' and list == ['a', 'n', 't']:
              list.append(j)
        elif j == 'n' and list == ['a', 'n', 't', 'o']:
              list.append(j)
    list = ''.join(list)


    if 'anton' in list:
        result.append(count)
    count += 1
print(*result)
```

## <center>–†–æ—Å–∫–æ–º–Ω–∞–≥–ª—è–¥ –∑–∞–±–æ—Ä–æ–Ω–∏–≤ –±—É–∫–≤—É –∞ üå∂Ô∏èüå∂Ô∏è</center>
–ù–µ–æ–±—Ö—ñ–¥–Ω–æ –Ω–∞–ø–∏—Å–∞—Ç–∏ –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ —Ä–µ–∞–ª—ñ–∑—É—î –∞–ª–≥–æ—Ä–∏—Ç–º –Ω–∞–ø–∏—Å–∞–Ω–Ω—è —Ü—ñ—î—ó –ø—ñ—Å–Ω—ñ. –ê–ª–≥–æ—Ä–∏—Ç–º –≤–∏–≤–æ–¥–∏—Ç—å –Ω–∞–ø—Ä–∏–∫—ñ–Ω—Ü—ñ —Ä–µ—á–µ–Ω–Ω—è –Ω–∞—Å—Ç—É–ø–Ω—É –≤ –∞–ª—Ñ–∞–≤—ñ—Ç–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É –±—É–∫–≤—É, —è–∫—â–æ –≤–æ–Ω–∞ –∑—É—Å—Ç—Ä—ñ—á–∞—î—Ç—å—Å—è —É —Ä—è–¥–∫—É —Ç–µ–∫—Å—Ç—É, –∞ —á–µ—Ä–≥–æ–≤–∏–π —Ä—è–¥–æ–∫ –≤—ñ–¥–æ–±—Ä–∞–∂–∞—î –≤–∂–µ –±–µ–∑ —Ü—ñ—î—ó –±—É–∫–≤–∏.

**–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö**

–ù–∞ –≤—Ö—ñ–¥ –ø—Ä–æ–≥—Ä–∞–º—ñ –ø–æ–¥–∞—î—Ç—å—Å—è –æ–¥–Ω–µ —Å–ª–æ–≤–æ, –∑–∞–ø–∏—Å–∞–Ω–µ –º–∞–ª–∏–º–∏ —Ä–æ—Å—ñ–π—Å—å–∫–∏–º–∏ –ª—ñ—Ç–µ—Ä–∞–º–∏ –±–µ–∑ –ª—ñ—Ç–µ—Ä–∏ "–µ".

**–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö**

–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∑–∞–∑–Ω–∞—á–µ–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É —Ä—è–¥–∫–∞, –∫—ñ–ª—å–∫—ñ—Å—Ç—å —è–∫–∏—Ö –¥–æ—Ä—ñ–≤–Ω—é—î –∫—ñ–ª—å–∫–æ—Å—Ç—ñ —Ä—ñ–∑–Ω–∏—Ö –ª—ñ—Ç–µ—Ä —É —Ä—è–¥–∫—É, —â–æ –≤–∏—Ö–æ–¥–∏—Ç—å —à–ª—è—Ö–æ–º –∫–æ–Ω–∫–∞—Ç–µ–Ω–∞—Ü—ñ—ó –≤–≤–µ–¥–µ–Ω–æ–≥–æ —Å–ª–æ–≤–∞ —Ç–∞ —Ä—è–¥–∫–∞ "–∑–∞–±–æ—Ä–æ–Ω–∏–≤ –±—É–∫–≤—É".

```python
word = input() + ' –∑–∞–ø—Ä–µ—Ç–∏–ª –±—É–∫–≤—É'
b = ['–∞', '–±', '–≤', '–≥', '–¥', '–µ', '–∂', '–∑', '–∏', '–π', '–∫', '–ª', '–º', '–Ω', '–æ', '–ø', '—Ä', '—Å', '—Ç', '—É', '—Ñ', '—Ö', '—Ü', '—á', '—à', '—â', '—ä', '—ã', '—å', '—ç', '—é', '—è']
for i in range(len(b)):
    if b[i] in word:
        print(word, b[i])
        word = word.replace(b[i], '')
        word = ' '.join(word.split())
```

## <center>–ó—Ä—É—à–µ–Ω–Ω—è —É —Ä–æ–∑–≤–∏—Ç–∫—É</center>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è —Ä—è–¥–æ–∫ —Ç–µ–∫—Å—Ç—É –∑ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–∏—Ö —á–∏—Å–µ–ª. –ó –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —Ä—è–¥–∫–∞ —Ñ–æ—Ä–º—É—î—Ç—å—Å—è –ø–µ—Ä–µ–ª—ñ–∫ —á–∏—Å–µ–ª. –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É —Ü–∏–∫–ª—ñ—á–Ω–æ–≥–æ –∑—Å—É–≤—É –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —Å–ø–∏—Å–∫—É –ø—Ä–∞–≤–æ—Ä—É—á, –∫–æ–ª–∏ –æ—Å—Ç–∞–Ω–Ω—ñ–π –µ–ª–µ–º–µ–Ω—Ç —Å—Ç–∞—î –ø–µ—Ä—à–∏–º, –∞ —Ä–µ—à—Ç–∞ –∑—Å—É–≤–∞—î—Ç—å—Å—è –Ω–∞ –æ–¥–Ω—É –ø–æ–∑–∏—Ü—ñ—é –≤–ø–µ—Ä–µ–¥, —É –±—ñ–∫ –∑–±—ñ–ª—å—à–µ–Ω–Ω—è —ñ–Ω–¥–µ–∫—Å—ñ–≤.

**–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö**

–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è —Ä—è–¥–æ–∫ —Ç–µ–∫—Å—Ç—É –∑ —Ä–æ–∑–¥—ñ–ª–µ–Ω–∏—Ö –ø—Ä–æ–±—ñ–ª–∞–º–∏ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–∏—Ö —á–∏—Å–µ–ª.

**–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö**

–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ –µ–ª–µ–º–µ–Ω—Ç–∏ –∑–º—ñ–Ω–µ–Ω–æ–≥–æ —Å–ø–∏—Å–∫—É —ñ–∑ —Ü–∏–∫–ª—ñ—á–Ω–∏–º –∑—Å—É–≤–æ–º, —Ä–æ–∑–¥—ñ–ª—è—é—á–∏ –π–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∏ –æ–¥–Ω–∏–º –ø—Ä–æ–±—ñ–ª–æ–º.

```python
a = input().split()
a.insert(0, a[-1])
del a[-1]
print(*a)
```
