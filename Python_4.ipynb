## <center>Вивести матрицю</center>
На вхід до програми подаються два натуральні числа n і m, кожне на окремому рядку — кількість рядків і стовпців у матриці. Далі вводяться самі елементи матриці - слова, кожне на окремому рядку; поспіль йдуть елементи спочатку першого рядка, потім другого, і т.д.

Напишіть програму, яка зчитує елементи матриці один за одним, виводить їх у вигляді матриці, виводить порожній рядок і знову ту ж матрицю, але вже помінявши місцями рядки зі стовпцями: перший рядок виводиться як перший стовпець, і так далі.

**Формат вхідних даних**

На вхід до програми подаються два числа n і m — кількість рядків і стовпців у матриці, далі йдуть n×m слів, кожне на окремому рядку.

**Формат вихідних даних**

Програма повинна вивести лічену матрицю, за нею порожній рядок, і ту саму матрицю, але помінявши місцями рядки зі стовпцями. Елементи матриці розділяти одним пропуском.

```python
n, m = int(input()), int(input())

matrix = [[0]*m for _ in range(n)]
for row in range(n):
    for col in range(m):
        matrix[row][col] = input()

for col in range(m):
    for row in range(n):
        print(matrix[row][col], end = ' ')
    print()
```

## <center>Слід матриці</center>
Після квадратної матриці називається сума елементів головної діагоналі. Напишіть програму, яка виводить слід заданої квадратної матриці.

**Формат вхідних даних**

На вхід до програми подається натуральне число n — кількість рядків і стовпців у матриці, потім елементи матриці (цілі числа) рядково через пробіл.

**Формат вихідних даних**

Програма має вивести одне число – слід заданої матриці.

```python
n = int(input())
matrix = []
for i in range(n):
    temp = [int(num) for num in input().split()]
    matrix.append(temp)

count = 0
for i in range(len(matrix)):
    count += matrix[i][i]

print(count)
```

## <center>Більше середнього</center>
Напишіть програму, яка виводить кількість елементів квадратної матриці в кожному рядку, більших за середній арифметичний елементів даного рядка.

**Формат вхідних даних**

На вхід до програми подається натуральне число n — кількість рядків і стовпців у матриці, потім елементи матриці (цілі числа) рядково через пробіл.

**Формат вихідних даних**

Програма повинна вивести n чисел — для кожного рядка кількість елементів матриці, більших за середній арифметичний елементів даного рядка.

```python
n = int(input())
matrix = []

for _ in range(n):
    row = [int(i) for i in input().split()]
    matrix.append(row)
    
for i in range(n):
    counter = 0
    average = sum(matrix[i]) / n
    for j in range(n):
        if matrix[i][j] > average:
            counter += 1
    print(counter)
```

## <center>Максимальний в області</center>
Напишіть програму, яка виводить максимальний елемент у заштрихованій області квадратної матриці.

**Формат вхідних даних**

На вхід до програми подається натуральне число n — кількість рядків і стовпців у матриці, потім елементи матриці (цілі числа) рядково через пробіл.

**Формат вихідних даних**

Програма повинна вивести одне число - максимальний елемент у області квадратної матриці.

```python
n = int(input())
matrix = []
for i in range(n):
    temp = [int(num) for num in input().split()]
    matrix.append(temp)

maxi = -100    
    
for i in range(len(matrix)):
    for j in range(len(matrix)):
        if i >= j and matrix[i][j] > maxi:
            maxi = matrix[i][j]

print(maxi)
```
